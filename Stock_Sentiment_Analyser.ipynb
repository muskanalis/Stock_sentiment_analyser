```python
!pip install -q textblob


from textblob import TextBlob


texts = [
    "Apple stock is soaring after strong earnings report!",
    "Elon Musks Bold Move Sends Tesla Shares Skyrocketing Overnight",
    "Apple’s Secret Innovation Could Disrupt Entire Tech Industry",
    "Amazon’s Supply Chain Crisis: What It Means for Global Markets",
    "Google’s AI Breakthrough Sparks Investor Frenzy",
    "Netflix Faces a Crossroads: Can It Survive the Streaming Wars?",
    "Facebook’s Future at Risk Amid New Privacy Regulations",
    "Nvidia’s Chip Dominance Threatened by Unexpected Competitor",
    "Tesla’s Latest Autopilot Update Divides Investors and Experts",
    "Apple’s Stock Tumbles After Surprise Earnings Miss",
    "Bitcoin’s Rollercoaster Ride Continues: What’s Next for Crypto Investors?",
]


for i, text in enumerate(texts):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity  #range from -1 (negative) to 1 (positive)
    if sentiment > 0:
        mood = "Positive"
    elif sentiment < 0:
        mood = "Negative"
    else:
        mood = "Neutral"
    print(f"Text {i+1}: {text}\nSentiment score: {sentiment:.2f} ({mood})\n")
```

1. here we first imported TextBlob


2. then inserted sample headlines and tweets


3. lastly we analyzed sentiment polarity for each text

```python
import matplotlib.pyplot as plt

scores = []
labels = []

for text in texts:
    blob = TextBlob(text)
    scores.append(blob.sentiment.polarity)
    labels.append(text[:30] + "...")  # Short label

plt.bar(labels, scores, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.ylabel("Sentiment Polarity")
plt.title("Stock Sentiment Scores")
plt.show()
```
